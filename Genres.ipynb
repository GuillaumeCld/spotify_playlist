```python
import pandas as pd
import matplotlib.pyplot as plt
import pickle 
import seaborn as sns
```

Dictonary with all genres present in the dataset

```python
with open("MyData/dict_genres.pkl", "rb") as file:
    dict_count_genres = pickle.load(file)
    
print("Number of different genres :",len(dict_count_genres))
```

```python
dict_new = dict(filter(lambda x: x[1] > 2, dict_count_genres.items()))
dict_genres_sorted = dict(sorted(dict_new.items(), key=lambda item: item[1], reverse=True))
print("Top 10 genres in the dataset : ")
for e in list(dict_genres_sorted.items())[:10] : 
    print(e)
```
